```python
import pandas as pd
import jieba
import re
```

```python
df = pd.read_csv('/Users/xiaosass/Desktop/爬虫接单/携程大众点评评论爬取tf-idf分析/ctrip_comment.csv', header=None)
```

```python
#csv文件转化成txt文档，方便jieba处理
txt = open('/Users/xiaosass/Desktop/爬虫接单/携程大众点评评论爬取tf-idf分析/ctrip_comment.txt',
           mode='w')
for i in range(len(df[0])):
    txt.write(df[0][i]+',')
```

```python
#文本预处理，去除无用字符，只取出中文
with open('/Users/xiaosass/Desktop/爬虫接单/携程大众点评评论爬取tf-idf分析/ctrip_comment.txt') as f:
    data = f.read()
new_data = re.findall('[\u4e00-\u9fa5]+', data, re.S)
new_data = " ".join(new_data)
```

```python
#文本分词
seg_list_exact = jieba.lcut(new_data, cut_all=True)
object_list = []
#加载停用词
with open('/Users/xiaosass/Desktop/爬虫接单/携程大众点评评论爬取tf-idf分析/baidu_stopwords.txt',encoding='utf-8') as f1:
    con = f1.readlines()
    stop_words = set()
    for i in con:
        stop_words.add(i.strip())
#去掉停用词及单个字
for word in seg_list_exact:
    if word not in stop_words and len(word) >1:
        object_list.append(word)
object_list
```
