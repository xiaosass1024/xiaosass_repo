```python
import pandas as pd
import jieba
import re
```

```python
df = pd.read_csv('ctrip_comment.csv', header=None)
```

```python
#csv文件转化成txt文档，方便jieba处理
txt = open('ctrip_comment.txt',mode='w')
for i in range(len(df[0])):
    txt.write(df[0][i]+',')
```

```python
#文本预处理，去除无用字符，只取出中文
with open('ctrip_comment.txt') as f:
    data = f.read()
new_data = re.findall('[\u4e00-\u9fa5]+', data, re.S)
new_data = " ".join(new_data)
```

```python
#文本分词
seg_list_exact = jieba.lcut(new_data, cut_all=True)
object_list = []
#加载停用词
with open('baidu_stopwords.txt',encoding='utf-8') as f1:
    con = f1.readlines()
    stop_words = set()
    for i in con:
        stop_words.add(i.strip())
#去掉停用词及单个字
for word in seg_list_exact:
    if word not in stop_words and len(word) >1:
        object_list.append(word)
len(object_list)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#将文本中的词语转换为词频矩阵  
vectorizer = CountVectorizer(stop_words=None)
X = vectorizer.fit_transform(object_list)
bag_of_words = vectorizer.get_feature_names_out()
df_word = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
df_word
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
words_tfidf = {} #定义一个字典，用于后期排序
#实例化一个Tf-idf转化器
tfidf_transformer = TfidfTransformer()
#根据语料集的词袋向量计算TF-IDF(fit)
tfidf_transformer.fit(X.toarray())
for idx, word in enumerate(vectorizer.get_feature_names_out()):
#     print(f'{word}\t{tfidf_transformer.idf_[idx]}')
#数据写入字典，用于后面根据权重排序
    words_tfidf[word] = tfidf_transformer.idf_[idx]
#排序
words_tfidf_ordered = sorted(words_tfidf.items(), key=lambda x:x[1], reverse=True)
dict(words_tfidf_ordered)
```
