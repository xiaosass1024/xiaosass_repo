```python
import nltk
from nltk.collocations import BigramAssocMeasures
from nltk.collocations import TrigramAssocMeasures
from nltk.collocations import BigramCollocationFinder
from nltk.collocations import TrigramCollocationFinder
from nltk.corpus import genesis
```

```python
from nltk.corpus import stopwords
```

```python
bigram_measures = BigramAssocMeasures()
trigram_measures = TrigramAssocMeasures()
finder1 = BigramCollocationFinder.from_words(genesis.words('text.txt'))
finder1.apply_freq_filter(100)
finder1.nbest(bigram_measures.pmi, 100)
```

```python
finder2 = TrigramCollocationFinder.from_words(genesis.words('text.txt'))
finder2.apply_freq_filter(10)
finder2.nbest(trigram_measures.pmi, 100)
```
