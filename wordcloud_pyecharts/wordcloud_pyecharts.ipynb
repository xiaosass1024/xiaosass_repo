```python
import jieba
import collections
import re
from pyecharts.charts import WordCloud
from pyecharts.globals import SymbolType
from pyecharts import options as opts
from pyecharts.globals import ThemeType, CurrentConfig
```

```python
with open('/Users/xiaosass/Desktop/爬虫接单/携程评论爬取/wordcloud.txt') as f:
    data = f.read()
```

```python
new_data = re.findall('[\u4e00-\u9fa5]+', data, re.S)
new_data = ' '.join(new_data)
```

```python
seg_list_exact = jieba.cut(new_data, cut_all=True)
seg_list_exact
```

```python
result_list = list()
with open('/Users/xiaosass/Desktop/xiaosass_repo/wordcloud/baidu_stopwords.txt',encoding='utf-8') as f:
    con = f.readlines()
    stop_words = set()
    for i in con:
        i = i.replace("\n", '')
        stop_words.add(i)
```

```python
for word in seg_list_exact:
    if word not in stop_words and len(word)>1:
        result_list.append(word)
result_list
```

```python
word_conts = collections.Counter(result_list)
word_conts
```

```python
word_conts_top100 = word_conts.most_common(100)
word_conts_top100
```

```python
word1 = WordCloud(init_opts=opts.InitOpts(width='1350px', height='750px', theme=ThemeType.MACARONS))
word1.add('词频', data_pair=word_conts_top100, word_size_range=[15, 108], textstyle_opts=opts.TextStyleOpts(font_family='cursive'),shape=SymbolType.DIAMOND)
word1.set_global_opts(title_opts=opts.TitleOpts('酒店评论词云图'),
                     toolbox_opts=opts.TooltipOpts(is_show=True, background_color='red', border_color='yellow'))

word1.render("酒店评论词云图.html")
```
